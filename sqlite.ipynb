# SQL And SQLite
SQL(Structured Query Language) is a standard language for managing and manipulating relational databases. SQLite is a self-contained, serverless, and zero-configuration databases systems, In this lesson, we will cover the basics of SQL and SQLite, including creating databases tables, and performing various SQL operations.


SQLite is a lightweight version of SQL


```python
import sqlite3
```

```python
## Connect to a sqlite database
connection=sqlite3.connect('example.db')
connection
# In our folder we have created a 'example.db' file 
```

```python
cursor = connection.cursor()
```

```python
## Create a table 
cursor.execute('''
Create table if not exists Employees(
id Integer Primary Key,
name Text Not Null,
age Integer,
department Text)
''')

## Commit the changes 
connection.commit()
```

```python
cursor.execute('''
Select * from employees


''')
```

```python
## Insert the data in sqlite table 
cursor.execute('''
Insert Into employees(name,age,department)
               values('Manish',18,'Data Science')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Alex',28,'CyberSecurity')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Joan',24,'HR')
''')

## connection 
connection.commit()
```

```python
## Quering the data from the table 
cursor.execute('Select * from employees')
rows=cursor.fetchall()

## print the queried data 

for row in rows:
    print(row)
```

```python
## Update the data in the table

cursor.execute('''
UPDATE employees
Set age=34
where name="Manish"
''')

connection.commit()
```

```python
## Quering the data from the table 
cursor.execute('Select * from employees')
rows=cursor.fetchall()

## print the queried data 

for row in rows:
    print(row)
```

```python
## Delete the data from table 
cursor.execute('''
Delete from employees
              Where name ='Joan'
'''
)

connection.commit()
```
